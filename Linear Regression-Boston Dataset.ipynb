# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn.datasets import load_boston
import scipy.stats as stats
```

```python
bos = pd.DataFrame(boston.data, columns=boston.feature_names)
```

## Plotting graphs on Independent Variables

```python
plt.figure(figsize=(20,30), facecolor='white')
plotnumber = 1

for i in bos.columns:
    if plotnumber<14:
        ax=plt.subplot(5,3,plotnumber)
        plt.scatter(bos[i],bos['MEDV'])
        plt.xlabel(i,fontsize=15)
        plt.ylabel('MEDV', fontsize=15)
    plotnumber+=1
plt.tight_layout()
```

## Checking correlation between independent variables

```python
corre_mat = bos.corr().round(2)
```

```python
import seaborn as sns
sns.heatmap(data=corre_mat,annot=True)
```

```python
plt.figure(figsize=(20,5))

features = ['LSTAT','RM']
target = bos['MEDV']

for i,col in enumerate(features):
    plt.subplot(1,len(features),i+1)
    x = bos[col]
    y = target
    plt.scatter(x,y,marker='o')
```

```python
X = pd.DataFrame(np.c_[bos['LSTAT'],bos['RM']],columns=['LSTAT','RM'])
Y = bos['MEDV']
```

## Dividing dataset into train and test sets

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X, y, test_size = 0.2, random_state=5)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

## Building Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

lin_model = LinearRegression()
lin_model.fit(X_train, Y_train)
```

## Predicting on Training set

```python
y_pred = lin_model.predict(X_train)
```

```python
rmse = (np.sqrt(mean_squared_error(Y_train, y_pred)))
r2 = r2_score(Y_train, y_pred)
```

```python
print('RMSE score is={}'.format(rmse))
print('R2 score is={}'.format(r2))
```

## Predicting on Testing set

```python
y_test_pred = lin_model.predict(X_test)

rmse = (np.sqrt(mean_squared_error(Y_test, y_test_pred)))
r2 = r2_score(Y_test,y_test_pred)

print('RMSE score is={}'.format(rmse))
print('R2 score is={}'.format(r2))
```
